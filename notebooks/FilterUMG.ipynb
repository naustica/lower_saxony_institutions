```python
from google.cloud import bigquery
import pandas as pd
```

```python
client = bigquery.Client(project='subugoe-collaborative')
```

```python
filter_umg = client.query(f"""
                           SELECT inst.doi, address_full
                           FROM `subugoe-collaborative.openbib.kb_a_addr_inst` AS inst
                           JOIN `subugoe-collaborative.openbib.kb_inst_lookup` AS kb_inst
                               ON inst.inst_id_top = kb_inst.inst_id
                           WHERE kb_inst.ror = 'https://ror.org/01y9bpm73' 
                          """).to_dataframe()
```

```python
filter_umg.head()
```

```python
df = filter_umg.copy()
```

```python
df['institution'] = 'Universität Göttingen'
```

```python
# https://github.com/naustica/affiliation_mapping/blob/main/affiliation_mapping.py
# Universitätsmedizin
med = ['umg', 'oncol', 'med', 'cardiol', 'clin', 'pneumol', 'klini',
       'multiscale bioimaging', 'cardiovasc',
       'urol', 'gastroenter', 'surg', 'physio', 'psychia',
       'dent', 'pathol', 'radiol', 'adv.* imag', 'heilkunde',
       '\w*klin', 'robert[-\s]koch', 'hosp', 'otorhinolaryngol', 'otolaryngol',
       'immu', 'pharma[ck]ol', 'ophthal', 'rheuma']
```

```python
# https://ror.org/021ft0n22 UMG
```

```python
df.loc[df['address_full'].str.contains('|'.join(med), case=False, regex=True), ['institution']] = 'Universitätsmedizin Göttingen'
```
